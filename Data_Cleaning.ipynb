```python
# 1. Fill low-missing columns
# ---------------------------

# Courier Status
data["Courier Status"] = data["Courier Status"].fillna("Unknown")

# Currency → fill with most frequent value (mode)
currency_mode = data["currency"].mode()[0]
data["currency"] = data["currency"].fillna(currency_mode)

# Location columns
location_cols = ["ship-city", "ship-state", "ship-postal-code", "ship-country"]
for col in location_cols:
    data[col] = data[col].fillna("Unknown")

# Promotion IDs → missing means no promotion
data["promotion-ids"] = data["promotion-ids"].fillna("No Promotion")

# ---------------------------
# 2. Handle Amount column
# ---------------------------
# If order is Cancelled and Amount is missing → fill with 0
data.loc[(data["Status"] == "Cancelled") & (data["Amount"].isna()), "Amount"] = 0

# Optional: drop rows where Amount is still missing
data = data.dropna(subset=["Amount"])

# ---------------------------
# 3. Drop high-missing / useless columns
# ---------------------------
drop_cols = ["fulfilled-by", "Unnamed: 22"]
data = data.drop(columns=drop_cols, errors="ignore")

# ---------------------------
# 4. Quick validation
# ---------------------------
print("Remaining Missing Values:")
print(data.isna().sum())

print("\nFinal Columns:")
print(data.columns)

```

```python
data.to_csv("Amazon_Sale_Report_Cleaned.csv", index=False)
```

```python
import pandas as pd
import re
from rapidfuzz import process, fuzz
```

```python
# 1. Basic cleaning function
def basic_clean(city):
    city = str(city).lower()
    city = re.sub(r"\(.*?\)", "", city)              # remove brackets
    city = re.sub(r",.*", "", city)                  # remove after comma
    city = re.sub(r"\b(east|west|north|south)\b", "", city)  # remove directions
    city = re.sub(r"\d+", "", city)                  # remove numbers
    city = re.sub(r"[^\w\s]", "", city)              # remove symbols (. - / etc.)
    city = re.sub(r"\s+", " ", city).strip()         # normalize spaces
    return city.title()

data["city_clean"] = data["ship-city"].apply(basic_clean)

# 2. Master city list
master_cities = list(set(data["city_clean"]))

# 3. Fuzzy spelling correction
def correct_spelling(city):
    match, score, _ = process.extractOne(city, master_cities, scorer=fuzz.token_sort_ratio)
    if score > 90:   # you can tune this (85–95)
        return match
    else:
        return city

data["city_final"] = data["city_clean"].apply(correct_spelling)
```
